### About the Dataset:

The [Breast Cancer Diagnostic data](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29) is available on the UCI Machine Learning Repository. This database is also available through the [UW CS ftp server](http://ftp.cs.wisc.edu/math-prog/cpo-dataset/machine-learn/cancer/WDBC/).

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. n the 3-dimensional space is that described in: [K. P. Bennett and O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].

**Attribute Information**:

- ID number
- Diagnosis (M = malignant, B = benign) 3-32)

Ten real-valued features are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter) 
2. texture (standard deviation of gray-scale values) 
3. perimeter 
4. area 
5. smoothness (local variation in radius lengths) 
6. compactness (perimeter^2 / area - 1.0) 
7. concavity (severity of concave portions of the contour) 
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant

### Task 3: Classification using XGBoost (minimal feature selection)

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Basline model using XGBoost
x_train,x_test,y_train,y_test=train_test_split(df,y, test_size=0.3,random_state=42)

clf_1 = xgb.XGBClassifier(random_state=42) #Instantiated classfier
clf_1.fit(x_train,y_train)
```

```python
#Evaluation metrics
y_pred=clf_1.predict(x_test)
print('Accuracy is: ', accuracy_score(y_test,y_pred))
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='d')
```

We still have FP and FN as 2 which are risky and dangerous. ALso TP and TN ar ehigh resulting in higher accuracy of 97.6%

### Task 4: Univariate Feature Selection and XGBoost

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#SelectKBest - score it returns can be used to select top n features with their highest values for their chi2 
#statistics - measures stocastic measures from dataset
```

```python
select_feature  = SelectKBest(chi2,k=10) #selectes top 10 features
select_feature.fit(x_train,y_train)

#print features score
print('Score List: ',select_feature.scores_)
print('Feature List: ',x_train.columns)
```

```python
x_train_new=select_feature.transform(x_train) #trasnform data and only select tope 10 features
x_test_new=select_feature.transform(x_test)

clf_2=xgb.XGBClassifier().fit(x_train_new,y_train)
y_pred_new=clf_2.predict(x_test_new)
print('Accuracy is: ', accuracy_score(y_test,y_pred_new))
cm_new=confusion_matrix(y_test,y_pred_new)
sns.heatmap(cm_new,annot=True,fmt='d')
```

We still have FP (3) and FN as 2 which are risky and dangerous. ALso TP and TN are high resulting in higher accuracy of 97.06%

### Task 5: Recursive Feature Elimination with Cross-Validation

```python
# Pros - not only give best features but also optimal number of features through cross-validations

from sklearn.feature_selection import RFECV

clf_3=xgb.XGBClassifier()
rfecv=RFECV(estimator=clf_3,step=1,
           cv=5,scoring='accuracy',
           n_jobs=-1).fit(x_train,y_train) #step - how many features wanteed to eliminate at each step
#njobs - no. of cpus available

print('The no. of optimal features are: ', rfecv.n_features_)
print('Best features:\n', x_train.columns[rfecv.support_])
```

```python
print('The accuracy using RFECV is : ',accuracy_score(y_test,rfecv.predict(x_test)))
```

```python
num_features=[i for i in range(1,len(rfecv.grid_scores_)+1)]
cv_scores = rfecv.grid_scores_

ax=sns.lineplot(x=num_features,y=cv_scores)
ax.set(xlabel='No. of selected features',ylabel='CV scores')
```

### Task 6: Feature Extraction using Principal Component Analysis

```python
#to reduce feature space dimension, we can use PCA

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
#pre-processing step for PCA is normalizing it as PCA is affected by scale and we want all variables to be same
x_train_norm1=(x_train-x_train.mean())/(x_train.max()-x_train.min())
x_test_norm1=(x_test-x_test.mean())/(x_test.max()-x_test.min())
```

```python
from sklearn.decomposition import PCA

pca=PCA()
pca.fit(x_train_norm1)
#results from pca can be context dependent -cummulatively give you defined accuracy

plt.figure(1,figsize=(10,8))
sns.lineplot(data=np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No. of components')
plt.ylabel('Cummulative Explained variance')
```

can stop at 15 or 16 features as optimum value for high accuracy
